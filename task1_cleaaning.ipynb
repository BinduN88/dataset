```python
#data cleaning
import pandas as pd
Data=pd.read_csv("car_prices.csv")
print("Initial Dataset Preview:\n",Data.head())
```

```python
# Step 3: Check for missing values
print("\nMissing Values in Each Column:\n")
Data.isnull().sum()

```

```python
# Step 4: Drop missing values (optional, or use fillna())
Data.dropna() 
Data.fillna("unknown")
```

```python
# Step 5: Remove duplicate rows
Data= Data.drop_duplicates()
print(Data)
```

```python
# Step 6: Clean column names 
Data.columns = Data.columns.str.strip().str.lower().str.replace(" ", "_")
print(Data.columns)
```

```python
# Convert income column to float after removing $ and commas
if 'mmr' in Data.columns:
    Data['mmr'] = Data['mmr'].replace(r'[\$,]', '', regex=True).astype(float)
    print(Data['mmr'])
```

```python
# Step 8: Standardize text columns
if 'model' in Data.columns:
    Data['model'] = Data['model'].str.lower().str.strip()
    print( Data['model'])
```

```python
# Step 9: Final check of dataset
print("\nCleaned Dataset Preview:\n", Data.head())
print("\nUpdated Data Types:\n", Data.dtypes)
```

```python
# Step 10: Save the cleaned dataset
Data.to_csv("cleaned_dataset.csv", index=False)
print("\nCleaned dataset saved as 'cleaned_dataset.csv'")
```
